```python
from os import listdir
from sklearn.model_selection import train_test_split

def create_data():
    path = '../data/VCTK-Corpus/'
    path_audio = path + 'wav48/'
    path_label = path + 'txt/'
    X = []
    y = []
    
    for i, speaker in enumerate(listdir(path_audio)):
        speaker += '/'

        for wav in listdir(path_audio + speaker):
            txt = wav.replace('.wav', '.txt')
            X.append(path_audio + speaker + wav)
            y.append(path_label + speaker + txt)
            
        if i == 0:
            break
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
    
    with open('../data/VCTK-Corpus/train.csv', 'w') as train_file:
        for x, y in zip(X_train, y_train):
            train_file.write(x + ',' + y + '\n')
    with open('../data/VCTK-Corpus/test.csv', 'w') as test_file:
        for x, y in zip(X_test, y_test):
            test_file.write(x + ',' + y + '\n')
        
create_data()
```
